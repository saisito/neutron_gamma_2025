```python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
def show_two_images(image1_path, image2_path):
    """
    Given two image file paths, this function displays them side by side.
    
    Parameters:
    image1_path (str): Path to the first image.
    image2_path (str): Path to the second image.
    """
    # Check if the images exist
    if not os.path.exists(image1_path) or not os.path.exists(image2_path):
        print("Make sure the images exist at the provided paths.")
        return
    
    # Load the images
    image1 = imread(image1_path)
    image2 = imread(image2_path)
    
    # Set up the figure to display the images side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    # Display the first image
    axes[0].imshow(image1)
    axes[0].axis('off')
    
    # Display the second image
    axes[1].imshow(image2)
    axes[1].axis('off')
    
    # Adjust layout to prevent overlapping
    plt.tight_layout()
    plt.show()
```

#### Tresholds

```python
show_two_images('results_pure/thresholds_pure.png', 'results_salt/thresholds_salt.png')
```

#### Confusion Matrix

```python
show_two_images('results_pure/confusion_pure.png', 'results_salt/confusion_salt.png')
```

#### ROC AUC

```python
show_two_images('results_pure/roc_pure.png', 'results_salt/roc_salt.png')
```
